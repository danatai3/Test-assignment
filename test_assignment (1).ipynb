```python
import pandas as pd

# Загрузка данных
dataset_path = 'housing_prices.csv'
housing_data = pd.read_csv(dataset_path)

# Проверка данных
housing_data.head()
```

```python
import openai
from dotenv import load_dotenv
import os

# Загрузка токена OpenAI из .env файла
load_dotenv()
openai.api_key = os.getenv("openai_token")
```

```python
# Function to generate Instagram post
def generate_instagram_post(entry):
    prompt = f"Generate a real estate Instagram post for a property with price=${entry['price']} and area={entry['area']}."


    response = openai.Completion.create(
        engine="gpt-3.5-turbo",
        prompt=prompt,
        max_tokens=200
    )

    return response.choices[0].text.strip()

```

```python
# Generate EXAMPLE.txt
with open('/content/EXAMPLE.txt', 'w') as file:
    for _, row in housing_data.sample(20, replace=True).iterrows():
        file.write(f"price={row['Price']}\n")
        file.write(f"area={row['Area']}\n")
        file.write("// all parameters used to generate this sample\n")

        post_text = generate_instagram_post(row)
        file.write(post_text + '\n')
        file.write("------------------------ // divider\n")
```
